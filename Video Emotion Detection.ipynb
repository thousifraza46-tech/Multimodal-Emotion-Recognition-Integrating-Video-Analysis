```python
import cv2
from deepface import DeepFace

# Initialize video capture from webcam
video_capture = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()
    
    if not ret:
        break
    
    # Convert the frame from BGR to RGB (DeepFace requires RGB format)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    try:
        # Analyze emotions using DeepFace
        analysis = DeepFace.analyze(rgb_frame, actions=['emotion'], enforce_detection=False) 

        # Access the dominant emotion (check if faces are detected)
        dominant_emotion = analysis[0]['dominant_emotion'] 

        # Display the result on the frame
        cv2.putText(frame, dominant_emotion, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    except Exception as e:
        print("Error in emotion analysis:", e)

    # Show the video stream
    cv2.imshow('Emotion Detection', frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break 

# Release capture
video_capture.release()
cv2.destroyAllWindows()
```
